```python
#Load data
fPath = "/content/drive/MyDrive/Advent of Code/2021/Dec 05/"
# f = open(fPath+"input.txt", "r")
f = open(fPath+"input.txt", "r")
inputs = f.read()
lines = inputs.splitlines()
import pandas as pd
```

```python
#count total number of coordinates
numCoords = len(lines)
#split coordinates into start / end pairs
for l in range(0, len(lines)):
  lines[l] = lines[l].split(' -> ')
coords = pd.DataFrame(lines, columns={'start','end'})
#split coordinate pairs into individual numbers and cast to integers in new columns
coords[['startx', 'starty']] = coords['start'].str.split(',',expand=True)
coords[['endx', 'endy']] = coords['end'].str.split(',',expand=True)
coords['startx'] = pd.to_numeric(coords['startx'])
coords['starty'] = pd.to_numeric(coords['starty'])
coords['endx'] = pd.to_numeric(coords['endx'])
coords['endy'] = pd.to_numeric(coords['endy'])
coords['slope'] = (coords['endy'] - coords['starty']) / (coords['endx'] - coords['startx'])
coords
```

```python
# Part A
import numpy as np
#create a grid for keeping a count of lines per cell on grid
counter = np.zeros((max(coords['endx']) + 2,max(coords['endy']) + 2), dtype=int)
for i in range(0, numCoords):
  startx = coords.at[i, 'startx']
  starty = coords.at[i, 'starty']
  endx = coords.at[i, 'endx']
  endy = coords.at[i, 'endy']
  if coords.at[i, 'slope'] == 0: #if its a horizontal line, then increment counters along the horizontal line
    #print(str(i) + " -- " + str(startx) + "," + str(starty) + ' -> ' + str(endx) + "," + str(endy))
    for x in range(min(startx, endx), max(startx, endx) + 1):
      #print('Print: ' + str(x) + ',' + str(starty))
      counter[starty, x] += 1
  elif np.isinf(coords.at[i, 'slope']): #if its a vertical line, then increment counters
    #print(str(i) + " -- " + str(startx) + "," + str(starty) + ' -> ' + str(endx) + "," + str(endy))
    for y in range(min(starty, endy), max(starty, endy) + 1):
      #print('Print: ' + str(startx) + ',' + str(y))
      counter[y, startx] += 1
#don't increment any counters for a non-vertical or horizontal line
totalMoreThan2 = (counter>=2).sum()
```

```python
# Part B
counter = np.zeros((max(coords['endx']) + 2,max(coords['endy']) + 2), dtype=int)
for i in range(0, numCoords):
  livex = 0
  livey = 0
  startx = coords.at[i, 'startx']
  starty = coords.at[i, 'starty']
  endx = coords.at[i, 'endx']
  endy = coords.at[i, 'endy']
  if coords.at[i, 'slope'] == 0: #if its a horizontal line, then increment counters along the horizontal line
    #print(str(i) + " -- " + str(startx) + "," + str(starty) + ' -> ' + str(endx) + "," + str(endy))
    for x in range(min(startx, endx), max(startx, endx) + 1):
      #print('Print: ' + str(x) + ',' + str(starty))
      counter[starty, x] += 1
  elif np.isinf(coords.at[i, 'slope']): #if its a vertical line, then increment counters
    #print(str(i) + " -- " + str(startx) + "," + str(starty) + ' -> ' + str(endx) + "," + str(endy))
    for y in range(min(starty, endy), max(starty, endy) + 1):
      #print('Print: ' + str(startx) + ',' + str(y))
      counter[y, startx] += 1
  elif coords.at[i, 'slope'] == -1: #negative sloping (y decreasing as you move to higher x values)
    #print(str(i) + " -- " + str(startx) + "," + str(starty) + ' -> ' + str(endx) + "," + str(endy))
    for x in range(0, max(startx, endx) - min(startx, endx) + 1):
      livex = min(startx, endx) + x
      livey = max(starty, endy) - x
      #print('Print: ' + str(livex) + ',' + str(livey))
      counter[livey, livex] += 1
  elif coords.at[i, 'slope'] == 1: #positive sloping (y increasing as you move to higher x values)
    #print(str(i) + " -- " + str(startx) + "," + str(starty) + ' -> ' + str(endx) + "," + str(endy))
    for x in range(0, max(startx, endx) - min(startx, endx) + 1):
      livex = min(startx, endx) + x
      livey = min(starty, endy) + x
      #print('Print: ' + str(livex) + ',' + str(livey))
      counter[livey, livex] += 1
BtotalMoreThan2 = (counter>=2).sum()
```

```python
print('Part A answer is ' + str(totalMoreThan2))
print('Part B answer is ' + str(BtotalMoreThan2))
```
